```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ab=pd.read_pickle('../../Data/AB.pkl').sample(n=20000)
aab=pd.read_pickle('../../Data/AAB.pkl').sample(n=20000)
abb=pd.read_pickle('../../Data/ABB.pkl').sample(n=20000)
```

```python
colors=['DarkGray','Dodgerblue','Salmon']
names=[r'$\alpha\beta$',r'$\alpha\alpha\beta$',r'$\alpha\beta\beta$']
for i,df in enumerate([ab,aab,abb]):
    clonotypes=df.a1_V+df.a1_nSeqCDR3+df.b1_V+df.b1_nSeqCDR3
    if i==1:
        clonotypes=clonotypes+df.a2_V+df.a2_nSeqCDR3
    elif i==2:
         clonotypes=clonotypes+df.b2_V+df.b2_nSeqCDR3
    clone_size,counts=np.unique(np.unique(clonotypes,return_counts=1)[1],return_counts=1)
    plt.scatter(clone_size,counts/np.sum(counts),color=colors[i],label=names[i])
plt.xscale('log')
plt.yscale('log')
plt.ylim([1e-5,2])
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(frameon=False)
plt.show()
plt.close()
```
