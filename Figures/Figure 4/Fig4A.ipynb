```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn3_circles
```

```python
ab=pd.read_pickle('../../Data/AB.pkl')
aab=pd.read_pickle('../../Data/AAB.pkl')
abb=pd.read_pickle('../../Data/ABB.pkl')
```

```python
def repertoire_overlap(ab,aab,abb):
    """
    Venn diagrams for overlap of ab,aab, and abb amino acid repertoires.
    """
    for x in ['alpha','beta']:
        if x=='alpha': #Alpha single-chain repertoire
            a=np.array(ab.a1_aaSeqCDR3)
            b=np.hstack((np.array(aab.a1_aaSeqCDR3),np.array(aab.a2_aaSeqCDR3))) #include both alpha chains for aab 
            c=np.array(abb.a1_aaSeqCDR3)
        else: #Beta single-chain repertoire
            a=np.array(ab.b1_aaSeqCDR3)
            b=np.array(aab.b1_aaSeqCDR3)
            c=np.hstack((np.array(abb.b1_aaSeqCDR3),np.array(abb.b2_aaSeqCDR3))) #include both beta chains for abb 
        d=venn3_circles([set(np.unique(a)),set(np.unique(b)),set(np.unique(c))])
        colors=['Gray','Dodgerblue','Salmon']
        for i in range(3):
            d[i].set_lw(6)
            d[i].set_edgecolor(colors[i]) 
        plt.show()
        plt.close()
```

```python
repertoire_overlap(ab,aab,abb)
```
